## Handling missing values

```python
features_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>0]
```

```python
features_na
```

```python
for feature in features_na:
    
    print(f"The feature is '{feature}' \n and missing values percentage is  {np.round(dataset[feature].isnull().mean(), 4)}\n ")
    print('*'*42, '\n')
```

```python
# replacing categorical features

categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```
